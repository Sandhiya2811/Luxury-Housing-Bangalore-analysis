## Handle nulls in Unit_Size_Sqft, Ticket_Price_Cr, Amenity_Score

```python
house.isnull().sum()
```

```python
house[house['Unit_Size_Sqft'].isnull()]
```

```python
house.groupby(['Configuration'])['Unit_Size_Sqft'].median()
```

```python
def impute_Unit_Size_Sqft(row):

    if np.isnan(row['Unit_Size_Sqft']):

      if row['Configuration'] == "3BHK":

        return 5981.848326

      elif row['Configuration'] == "4BHK":
        return 5996.000000

      elif row['Configuration'] == "5BHK+":
        return 5988.000000
    else:
        return row['Unit_Size_Sqft']

house['Unit_Size_Sqft'] =  house.apply(impute_Unit_Size_Sqft, axis = 1)
```

```python
house.isnull().sum()
```

```python
house[house['Ticket_Price_Cr'].isnull()]
```

```python
house.groupby(['Configuration'])['Ticket_Price_Cr'].mean()
```

```python
def impute_Ticket_Price_Cr(row):

    if np.isnan(row['Ticket_Price_Cr']):

      if row['Configuration'] == "3BHK":

        return 12.662261

      elif row['Configuration'] == "4BHK":
        return 12.749241

      elif row['Configuration'] == "5BHK+":
        return 12.745952
    else:
        return row['Ticket_Price_Cr']

house['Ticket_Price_Cr'] =  house.apply(impute_Ticket_Price_Cr, axis = 1)
```

```python
house.isnull().sum()
```

```python
house[house['Amenity_Score'].isnull()]
```

```python
def impute_Amenity_Score(row):

    if np.isnan(row['Amenity_Score']):

      if row['Developer_Name'] == "Brigade":
        return 7.486064

      elif row['Developer_Name'] == "Embassy":
        return 7.481968

      elif row['Developer_Name'] == "Godrej":
        return 7.545401
    
      elif row['Developer_Name'] == "L&T Realty":
        return 7.516065

      elif row['Developer_Name'] == "Prestige":
        return 7.482133

      elif row['Developer_Name'] == "Puravankara":
        return 7.458331
    
      elif row['Developer_Name'] == "RMZ":
        return 7.491426

      elif row['Developer_Name'] == "SNN Raj":
        return 7.520361
    
      elif row['Developer_Name'] == "Sobha":
        return 7.527468

      elif row['Developer_Name'] == "Tata Housing":
        return 7.494180

      elif row['Developer_Name'] == "Total Environment":
        return 7.514738
    
    else:
        return row['Amenity_Score']

house['Amenity_Score'] =  house.apply(impute_Amenity_Score, axis = 1)
```

```python
import os
import psycopg2
import pandas as pd
from sqlalchemy import create_engine
from dotenv import load_dotenv

load_dotenv()

RENDER_DB_HOST = os.getenv("RENDER_DB_HOST")
RENDER_DB_NAME = os.getenv("RENDER_DB_NAME")
RENDER_DB_USER = os.getenv("RENDER_DB_USER")
RENDER_DB_PASSWORD = os.getenv("RENDER_DB_PASSWORD")
RENDER_DB_PORT = os.getenv("RENDER_DB_PORT")

# Connection string
DB_CONFIG = {
    'host': RENDER_DB_HOST,
    'dbname': RENDER_DB_NAME,
    'user': RENDER_DB_USER,
    'password': RENDER_DB_PASSWORD,
    'port': RENDER_DB_PORT
}

# Test psycopg2 connection
try:
    conn = psycopg2.connect(**DB_CONFIG)
    cursor = conn.cursor()
    cursor.execute("SELECT version();")
    print("✅ PostgreSQL Render connection successful!")
    print("Server version:", cursor.fetchone()[0])
    cursor.close()
    conn.close()
except Exception as e:
    print("❌ Error:", e)
```

# Run initial SQL validation queries

## SELECT COUNT(*)

```python
query = """
SELECT COUNT(*) AS total_row
FROM luxury_house_data;
"""

count = pd.read_sql(query, engine)
print(count)
```

## GROUP BY booking status

```python
import pandas as pd

df_cols = pd.read_sql("SELECT * FROM luxury_house_data;", engine)
print(df_cols.columns)
```

```python
query = """
SELECT "Booking_Status", COUNT(*) AS total_per_status
FROM luxury_house_data
GROUP BY "Booking_Status";
"""
book_status = pd.read_sql(query, engine)
print(book_status)
```

## AVG ticket price per builder

```python
query = """SELECT "Developer_Name", AVG("Ticket_Price_Cr") AS avg_ticket_price
FROM luxury_house_data
GROUP BY "Developer_Name"
ORDER BY avg_ticket_price DESC;
"""

avg_ticket = pd.read_sql(query, engine)
print(avg_ticket)
```

# 10th scenario


```python
query = """
SELECT 
    "Developer_Name",
    SUM("Ticket_Price_Cr") AS Total_Revenue,
    COUNT(*) AS Total_Booked
FROM 
    luxury_house_data
WHERE 
    "Booking_Status" = 1
GROUP BY 
    "Developer_Name"
ORDER BY 
    Total_Revenue DESC
LIMIT 5;
"""

top5_builders = pd.read_sql(query, engine)
print(top5_builders)
```
